# BLADE Autograd Demo


```python
import sys
sys.path.insert(0, ".")

import blade
import blade.nn   as nn
import blade.optim as optim
import math

print("BLADE imported successfully")
```

## 1. Creating Tensors


```python
x = blade.Tensor([3], requires_grad=True)
x.storage()[0]

a = blade.Tensor.from_data([3], [1.0, 2.0, 3.0])
a.requires_grad = True

print("shape :", a.shape)
print("values:", a.storage())
print("tracks gradients:", a.requires_grad)
print("is leaf node:", a.is_leaf)
```

## 2. Single-Op Gradient


```python
a = blade.Tensor.from_data([3], [1.0, 2.0, 3.0])
a.requires_grad = True

loss = blade.ops.sum(a * a)
print("loss value:", loss.item())

loss.backward()

print("grad (expected [2, 4, 6]):", a.grad.storage())
```

## 3. Multi-Op Chain


```python
a = blade.Tensor.from_data([2], [0.0, 1.0])
a.requires_grad = True

loss = blade.ops.sum(blade.ops.exp(a * 2.0))

loss.backward()

expected = [2 * math.exp(2 * x) for x in [0.0, 1.0]]
print("grad         :", a.grad.storage())
print("expected     :", [round(x, 5) for x in expected])
```
